```python
# Importing all necessary libraries
import cv2
import os
from PIL import Image
from pytesseract import pytesseract
import requests

import sys
import shutil
```

```python
def video_download(url):

    chunk_size = 256
    r = requests.get(url, stream = True)

    with open('video_1.mp4','wb') as f:
        for chunk in r.iter_content(chunk_size = chunk_size):
            f.write(chunk)

    print('Video downloaded successfully')
```

```python
def data_directory_creation():
    
    file_path = r"C:/Users/purva/OneDrive/Desktop/Data Science MSc/Dissertation/video_1.mp4"
    
    # Read the video from specified path
    cam = cv2.VideoCapture(file_path)

    try:

        # creating a folder named data
        if not os.path.exists('data'):
            os.makedirs('data')

    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')

    # frame
    currentframe = 0

    while(True):

        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = './data/frame' + str(currentframe) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
        else:
            break

    # Release all space and windows once done
    cam.release()
    cv2.destroyAllWindows()

    #Define path to tessaract.exe
    path_to_tesseract = r'C:\Users\purva\anaconda3\T\tesseract.exe'

    #Define path to images folder
    path_to_images = r"C:/Users/purva/OneDrive/Desktop/Data Science MSc/Dissertation/data/"

    #Point tessaract_cmd to tessaract.exe
    pytesseract.tesseract_cmd = path_to_tesseract

    data_list=[]

    #Get the file names in the directory
    for root, dirs, file_names in os.walk(path_to_images):
        #Iterate over each file name in the folder
        for file_name in file_names:
            #Open image with PIL
            img = Image.open(path_to_images + file_name)

            #Extract text from image
            text = pytesseract.image_to_string(img)

            data_list.append(text)
    #         print(text)

    final_data1 = set(data_list)
    
    return final_data1
```

```python
def delete_video():
    if os.path.exists("video_1.mp4"):
        os.remove("video_1.mp4")
        print('Video deleted successfully')
    else:
        print("The file does not exist")

def data_directory_deletion():
    file_name="data"
    shutil.rmtree(file_name)
    print('Data folder deleted successfully')
```
