```python
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm
import json
```

```python
#Load a new spacy model.
nlp_object = spacy.blank("en")

#Create a DocBin object
docbin_object = DocBin() 
```

```python
#Training data.
annotations_file = open('annotations.json',encoding="utf8")
TRAIN_DATA = json.load(annotations_file)
```

```python
TRAIN_DATA
```

```python
for text, annot in tqdm(TRAIN_DATA['annotations']): 
    doc = nlp_object.make_doc(text) 
    ents = []
    for start, end, label in annot["entities"]:
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    ents
    doc.ents = ents
    doc.ents
    docbin_object.add(doc)

docbin_object.to_disk("./training_data.spacy") # save the docbin object
```
