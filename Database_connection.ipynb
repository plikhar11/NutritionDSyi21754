```python
def nutri_val_extract_list(ing_list):
    serving_wt = []
    brand_name = []
    cal = []
    carbs = []
    pro = []
    fat= []
    sugar = []
    fib = []
    for ingredient_name in ing_list:
        ingredient_name1 = ingredient_name.replace(' ','&')
        BASE_URL = "https://api.nutritionix.com/v1_1/search/"+ingredient_name1+"?results=0:1&fields=nf_serving_weight_grams,item_name,nf_calories,nf_total_carbohydrate,nf_protein,nf_sugars,nf_total_fat,nf_dietary_fiber&item_type=3&appId="+APIID+"&appKey="+APIKEY
        response = requests.get(BASE_URL).json()
        #print(response)
        if not response['hits']: #if ingredient not found
#             export_df['Ingredient_name']=ingredient_name
            serving_wt.append(0)
            brand_name.append(0)
            cal.append(0)
            carbs.append(0)
            pro.append(0)
            fat.append(0)
            sugar.append(0)
            fib.append(0)
        else:
#             export_df['Ingredient_name']=ingredient_name
            serving_wt.append(response['hits'][0]['fields']['nf_serving_weight_grams'])
            brand_name.append(response['hits'][0]['fields']['item_name'])
            cal.append(response['hits'][0]['fields']['nf_calories'])
            carbs.append(response['hits'][0]['fields']['nf_total_carbohydrate'])
            pro.append(response['hits'][0]['fields']['nf_protein'])
            fat.append(response['hits'][0]['fields']['nf_total_fat'])
            sugar.append(response['hits'][0]['fields']['nf_sugars'])
            fib.append(response['hits'][0]['fields']['nf_dietary_fiber'])
            
    return ing_list,serving_wt,brand_name,cal,carbs,pro,fat,sugar,fib
```
